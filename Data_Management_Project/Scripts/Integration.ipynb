## CREAZIONE DATAFRAME PILOTI

Importazione del dataset driver e pulizia

```python
import csv
import pandas as pd
from datetime import datetime

# importazione dataframe driver
driver = pd.read_csv("Dataset F1/driver.csv", sep = ',', header = None)
driver.columns = ['driverId', 'driverRef', 'number', 'code', 'forename', 'surname', 'dob', 'nationality', 'url']
driver = driver.drop(['driverRef', 'url'], axis = 1)
driver['dob'] = pd.to_datetime(driver['dob'], format = '%Y-%m-%d')
driver
```

Importazione del dataset driver_standings e pulizia

```python
# importazione dataframe driver_standings
driver_standings = pd.read_csv("Dataset F1/driver_standings.csv", sep = ',', header = None)    
driver_standings.columns = ['driverStandingsId', 'raceId', 'driverId', 'seasonPoints', 'standingPosition', 'positionText', 'seasonWins']
driver_standings = driver_standings.drop(['driverStandingsId', 'positionText'], axis = 1)
driver_standings
```

Unione del dataset contenente i driver con standings

```python
# creazione dataframe informazioni piloti (unione dei due df precedenti)
driver_merge = pd.merge(driver, driver_standings, on = "driverId")
driver_merge['name'] = driver_merge['forename'] + ' ' + driver_merge['surname'] 
driver_merge = driver_merge.drop(['forename', 'surname'], axis = 1)
driver_merge = driver_merge[['driverId', 'name', 'number', 'code', 'dob', 'nationality', 'raceId', 'seasonPoints', 'standingPosition', 'seasonWins']]
driver_merge = driver_merge[driver_merge['raceId'] == 1020]
driver_merge = driver_merge.reset_index(drop = True)
driver_merge
```

Importazione del dataset results e pulizia

```python
# creazione dataframe results
results = pd.read_csv("Dataset F1/results.csv", sep = ',', header = None)
results = results.drop([0, 4, 7, 8, 12, 13, 14, 15, 16], axis = 1)
results.columns = ['raceId', 'driverId', 'constructorId', 'startPosition', 'finalPosition', 'racePoints', 'laps', 'time', 'statusId']
results = results[results['raceId'] == 1020]
results = results.drop('raceId', axis = 1)
results
```

Inserimento dei risultati in driver_merge

```python
# unione in driver_merge
driver_merge = pd.merge(driver_merge, results, on = "driverId")
driver_merge
```

Importazione del dataset status contenente tutti i possibili stati del pilota al termine della gara

```python
# creazione dataframe status
status = pd.read_csv("Dataset F1/status.csv", sep = ',', header = None)
status.columns = ['statusId', 'status']
status
```

Inserimento degli stati in driver_merge

```python
# unione in driver_merge
driver_merge = pd.merge(driver_merge, status, on = "statusId")
driver_merge = driver_merge.drop('statusId', axis = 1)
driver_merge
```

**Calcolo inizio gara**

Viene ricercato il primo tweet del canale ufficiale durante la gara.

Partendo dall'orario del tweet viene ricavato l'orario di inizio gara utilizzando i tempi dei piloti nei giri precedenti

```python
# ricerca del primo tweet effettuato dal canale ufficiale
for i in range(0, len(lap_dict)):
    if(lap_dict[i]['Tweet official channel'] != None):
        lap_temp = i
        time_temp = lap_dict[i]['Tweet official channel']['created_at']
        break
print(lap_temp)
# orario pubblicazione tweet
print(time_temp)
```

```python
# calcolo del tempo passato dall'inizio della gara alla pubblicazione del tweet
# tramite l'ausilio del tempo dei piloti
import datetime
tempo = 0
for i in range(0, lap_temp):
    for j in range(0, len(lap_dict[i]['Classifica'])):
        if lap_dict[i]['Classifica'][j]['Posizione'] == 1:
            tempo += lap_dict[i]['Classifica'][j]['Millisecondi']

tempo += 90000 # 1.30 minuti in millisecondi: tempo arbitrario dall'inizio del giro alla publicazione del tweet

tempo = datetime.datetime.fromtimestamp(tempo/1000.0)
inizio_gara = tempo.replace(year = time_temp.year,
                            month = time_temp.month,
                            day = time_temp.day,
                            hour = time_temp.hour - tempo.hour, 
                            minute = time_temp.minute - tempo.minute,
                            second = time_temp.second - tempo.second,
                            microsecond = 0,
                            tzinfo = time_temp.tzinfo)
# orario di inizio gara
inizio_gara
```

Calcolo tempi inizio (primo pilota) e fine giro (ultimi pilota)

```python
import datetime
from datetime import timedelta

def sum_hour(t1, t2):
    t1 = t1.replace(year = t2.year,
                    month = t2.month,
                    day = t2.day,
                    tzinfo = t2.tzinfo)
    t_sum = t1 + timedelta(hours = t2.hour,
                           minutes = t2.minute,
                           seconds = t2.second,
                           microseconds = t2.microsecond)
    return t_sum
    
for i in range(0, len(lap_dict)):
    for j in range(0, len(lap_dict[i]['Classifica'])):
        # millisecondi primo pilota
        if lap_dict[i]['Classifica'][j]['Posizione'] == 1:
            f = lap_dict[i]['Classifica'][j]['Millisecondi']
        # millisecondi ultimo pilota
        if isinstance(lap_dict[i]['Classifica'][j]['Posizione'], int) and lap_dict[i]['Classifica'][j]['Posizione'] == max(lap_dict[i]['Classifica'][m]['Posizione'] for m in range(0, len(lap_dict[i]['Classifica'])) if isinstance(lap_dict[i]['Classifica'][m]['Posizione'], int)):
            l = lap_dict[i]['Classifica'][j]['Millisecondi']
    f = datetime.datetime.fromtimestamp(f/1000.0)
    l = datetime.datetime.fromtimestamp(l/1000.0)
    if lap_dict[i]['Lap'] == 1:
        lap_dict[i]['Inizio giro'] = inizio_gara
        lap_dict[i]['Fine giro'] = sum_hour(l, inizio_gara)
    else:
        lap_dict[i]['Inizio giro'] = sum_hour(f, lap_dict[i-1]['Inizio giro'])
        lap_dict[i]['Fine giro'] = sum_hour(l, lap_dict[i-1]['Fine giro'])      
```
